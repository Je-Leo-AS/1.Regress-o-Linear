```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
```

```python
def reg_lin(points):
    x1 = points[:,0]
    y1 = points[:,1]
    x2 = x1**2
    y2  = y1**2
    xy = x1*y1
    n = len(points)
    a = (n*np.sum(xy)-np.sum(x1)*np.sum(y1))/(n*np.sum(x2)-(np.sum(x1))**2)
    b = (np.sum(y1) - a*np.sum(x1))/n
    return a,b
```

```python
points = np.array([[0.0,0.20],
                   [0.1,0.30],
                   [0.2,0.45],
                   [0.3,0.70],
                   [0.4,0.80]])
x , y = sp.symbols(['x','y'])
a, b = reg_lin(points)
eq = a*x + b
eq
```

```python
fig = plt.figure()
ax = plt.axes()
xs = points[:,0]
ys = points[:,1]
ax.plot(xs, xs*a+b)
ax.scatter(xs,ys)
```
